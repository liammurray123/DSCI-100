```r
ufo_data <- read_csv("UFOs_coord.csv")
# ufo_data
ufo_data_by_country <- arrange(ufo_data, Country)
options(repr.plot.width = 20, repr.plot.height =5)
ufo_plot <- ggplot(ufo_data_by_country, aes(x=State, color=Country)) + geom_histogram(stat="count", width=0.5) +
xlab("US State / Canadian Province") + ylab("Number of reports")
ufo_plot
```
